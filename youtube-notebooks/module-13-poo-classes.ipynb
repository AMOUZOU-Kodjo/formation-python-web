```python
class CompteBancaire:    def __init__(self, titulaire, solde=0):        self.titulaire = titulaire        self.solde = solde        self._historique = []    def deposer(self, montant):        self.solde += montant        self._historique.append(f"Dépôt : {montant}€")    def retirer(self, montant):        if montant <= self.solde:            self.solde -= montant            self._historique.append(f"Retrait : {montant}€")        else:            print("Solde insuffisant")    def afficher_solde(self):        print(f"Solde de {self.titulaire} : {self.solde}€")compte = CompteBancaire("Alice", 1000)compte.deposer(500)compte.retirer(200)compte.afficher_solde()  # Solde de Alice : 1300€
```
